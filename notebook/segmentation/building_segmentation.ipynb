```python
import os
import numpy as np
import random
from PIL import Image
from cmap import cmap 

import torch
import cv2
from transformers import pipeline
import matplotlib.pyplot as plt
```

```python
segmenter = pipeline('image-segmentation', 'nvidia/segformer-b0-finetuned-ade-512-512')
```

```python
in_path = '../../data/raw/Hubballi_Original/'
out_path = '../../data/processed/Hubballi_Processed/buildings/'
data_list = os.listdir(in_path)
random.seed(321)
# data_list = random.sample(data_list, 10)
```

```python
def segment_image(in_file, out_file):
    input_image = Image.open(in_file)
    results = segmenter(input_image)
    for mask in results:
        if mask['label'] == 'building':
            mask['mask'].save(out_file)
```

```python
for file in data_list:
    in_file = in_path + file
    out_file = out_path + file
    segment_image(in_file, out_file)
    print(f"{file} successfully segmented.")
```
