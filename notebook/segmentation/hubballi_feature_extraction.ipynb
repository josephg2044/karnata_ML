```python
import os
import numpy as np
import pandas as pd
import random
from PIL import Image
from features import features 

import torch
import cv2
from transformers import pipeline
import matplotlib.pyplot as plt
```

```python
segmenter = pipeline('image-segmentation', 'nvidia/segformer-b0-finetuned-ade-512-512')
```

```python
ignored = {"filter"}
in_path = '../../data/raw/Hubballi_Original/'
data_list = [x for x in os.listdir(in_path) if x not in ignored]
random.seed(321)
# data_list = random.sample(data_list, 10)
data_list[:10]
```

```python
df = pd.DataFrame(columns=['sl_no']+features)
```

```python
def extract_features(filename):
    return [feature['label'] for feature in segmenter(Image.open(filename))]
```

```python
for i in range(len(data_list)):
    filename = data_list[i]
    img_features = extract_features(in_path + filename)
    sl_no = filename[:-4]
    df.loc[i] = [int(sl_no)] + [int(feature in img_features) for feature in features]
    print(f"Image ({i + 1}, {sl_no}) segmentation completed")
df
```

```python
df.loc[len(df)] = ['avg'] + [df[feature].mean() for feature in features]
df
```
